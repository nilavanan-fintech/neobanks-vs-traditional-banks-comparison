# Week 4 — Product-depth score

Reuses the Week 3 Yes/Partial/No feature-matrix scoring and normalizes it to a 0–100
scale so it can feed the composite index alongside the other 0–100 dimensions.

**Input:** `data/processed/feature_matrix_scored.csv`

**Output:** `data/processed/product_depth_scores.csv`

```python
from pathlib import Path

import pandas as pd

ROOT = Path.cwd().parent
PROCESSED = ROOT / "data" / "processed"

df = pd.read_csv(PROCESSED / "feature_matrix_scored.csv")
n_features = len(df) - 1  # last row is the TOTAL row
totals = df[df["feature"].str.startswith("TOTAL")].iloc[0]

jupiter_total = float(totals["jupiter_score"])
hdfc_total = float(totals["hdfc_bank_score"])
print(f"features scored: {n_features}")
print(f"Jupiter total: {jupiter_total}/{n_features}")
print(f"HDFC Bank total: {hdfc_total}/{n_features}")
```

```python
scores = pd.DataFrame([
    {"entity": "Jupiter", "features_scored": n_features, "raw_score": jupiter_total, "product_depth_score": round(jupiter_total / n_features * 100, 1)},
    {"entity": "HDFC Bank", "features_scored": n_features, "raw_score": hdfc_total, "product_depth_score": round(hdfc_total / n_features * 100, 1)},
])
scores.to_csv(PROCESSED / "product_depth_scores.csv", index=False)
print("-> data/processed/product_depth_scores.csv")
scores
```
